```python
import pandas as pd
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split

from keras.models import load_model
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
eap_model = KerasClassifier(build_fn=load_model('rnn/eap_model.h5'))
hpl_model = KerasClassifier(build_fn=load_model('rnn/hpl_model.h5'))
mws_model = KerasClassifier(build_fn=load_model('rnn/mws_model.h5'))
```

```python
train = pd.read_csv('train.csv')
targets = ['EAP', 'HPL', 'MWS']
print(train.head())
```

```python
x = train['text']
y = train['author']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
char_voting = VotingClassifier(estimators=[('eap', eap_model), ('hpl', hpl_model), 
                                           ('mws', mws_model)], voting='hard')

char_voting = char_voting.fit(x, y)

print(char_voting.predict(x))
```
